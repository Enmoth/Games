```python
df = pd.read_csv('games.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df
```

```python
df = df.drop('Backlogs', axis=1)
```

```python
df
```

```python
def convert_value(value):
    if 'k' in value.lower():
        return float(value.lower().replace('k', '')) * 1000
    else:
        try:
            return float(value)
        except ValueError:
            return value
```

```python
df['Plays'] = df['Plays'].apply(convert_value)
```

```python
df
```

```python
df['Playing'] = df['Playing'].apply(convert_value)
df['Wishlist'] = df['Wishlist'].apply(convert_value)
```

```python
df['Times Listed'] = df['Times Listed'].apply(convert_value)
df['Number of Reviews'] = df['Number of Reviews'].apply(convert_value)
```

```python
df
```

```python
# Calcular a soma total de plays para cada gênero
genre_counts = df.groupby('Genres')['Plays'].sum().reset_index()

# Ordenar o DataFrame em ordem decrescente com base nas somas de plays
genre_counts_sorted = genre_counts.sort_values('Plays', ascending=False)
```

```python
print(genre_counts_sorted)
```

```python
df_genre_mix = genre_counts_sorted
```

```python
df_genre_mix
```

```python
genre_counts = df['Genres'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']
genre_counts_sorted = genre_counts.sort_values('Count', ascending=False)

df_genre_mix_count = genre_counts_sorted

print(genre_counts_sorted)
```

```python
df_copy = df.copy()

df_copy['Genres'] = df_copy['Genres'].str.split(',')
df_copy = df_copy.explode('Genres')
df_copy['Genres'] = df_copy['Genres'].str.replace(r"[\[\]']", "").str.strip()
genre_counts = df_copy['Genres'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']
genre_counts['Total Plays'] = df_copy.groupby('Genres')['Plays'].sum().values
genre_counts_sorted = genre_counts.sort_values('Total Plays', ascending=False)

df_genre_plays = genre_counts_sorted

print(genre_counts_sorted)
```
